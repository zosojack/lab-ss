## Parte 0: nomi file e funzioni

```python
# Nomi file 
# orangeF sta per filtro arancione
campione_1 = 'C240830_'
campione_2 = 'C240920_'
d_o_emi = ['0', '0o5', '1', '1o5', '2', '2o5', '3']#, '3o5']
d_o_ass = ['0', '1']
arr_spot = [1, 2, 3]
```

## CAMPIONE 1 REFERENCE

```python
raccoglitore_1 = {
    'spot1': {
        'ass=0': {
            'emi=0o5':  [],
            'emi=1':    [],
            'emi=1o5':  [],
            'emi=2':    [],
            'emi=2o5':  [],
            'emi=3':    []
        },
        'ass=1': []
    },
    'spot2': {
        'ass=0': {
            'emi=0o5':  [],
            'emi=1':    [],
            'emi=1o5':  [],
            'emi=2':    [],
            'emi=2o5':  [],
            'emi=3':    []
        },
        'ass=1': []
    },
    'spot3': {
        'ass=0': {
            'emi=0o5':  [],
            'emi=1':    [],
            'emi=1o5':  [],
            'emi=2':    [],
            'emi=2o5':  [],
            'emi=3':    []
        },
        'ass=1': []
    }
}
```

```python
for ass in d_o_ass:
    emi_list = d_o_emi[1:] if ass == '0' else ['0']  # Per ass=1 usa solo emi='0'
    
    # Per ciascuno degli emi > 0 se ass=0, solo emi=0 se ass=1
    for emi in emi_list:
        arr_energia = []

        # Ciclo sugli spot
        for spot in arr_spot:
            try:
                nome_file = spot_name(campione_1, emi, ass, spot)
                asse_x, asse_y = leggi_file(nome_file)
                print(min(asse_x), ',', max(asse_x))
            except FileNotFoundError:
                print(f"File non trovato: {nome_file}")
                continue
            
            picchi = find_peak_in_regions (asse_x, asse_y)
            
            fit_result = fit_gaussiani (asse_x, asse_y, picchi, emi, ass, spot, 5)
            
            if ass == '1':
                raccoglitore_1['spot'+str(spot)]['ass=1'].append(fit_result)
            else:
                raccoglitore_1['spot'+str(spot)]['ass=0']['emi='+emi].append(fit_result)
```
